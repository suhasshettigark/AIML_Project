```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('https://talentcocomedia.s3.amazonaws.com/ml-assets/titanic.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
data=df.copy()
```

```python
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(missing_values=np.nan,strategy="mean")

imputer=imputer.fit(data[['Age']])

data['Age']=imputer.transform(data[['Age']])
```

```python
data.info()
```

```python
data=df.fillna({'Age':df['Age'].mean()})
```

```python
data.isnull().sum(axis=0)
```

```python
data=data.drop(['Cabin'],axis=1)
```

```python
data.head()
```

```python
data=data.dropna()
```

```python
data.info()
```

```python
data=data.drop(['PassengerId','Name','Ticket','Embarked'],axis=1)
```

```python
data.head()
```

**Day 5**

---

**Using LogisticRegression**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(penalty='l2', solver= 'newton-cg', C=5)
```

```python
lr_model.fit(x_train,y_train)
```

```python
y_pred_lr=lr_model.predict(x_test)
y_pred_lr
```

```python
from sklearn.metrics import accuracy_score
acc_lr=accuracy_score(y_pred_lr,y_test)
acc_lr
```

Day 6

**Using KNN classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
```

```python
neigh.fit(x_train, y_train)
```

```python
y_pred_knn = neigh.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred_knn)
accuracy
```

**Using Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(x_train,y_train)
```

```python
y_pred_dtc = clf.predict(x_test)
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred_dtc)
accuracy
```
